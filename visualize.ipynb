```python
import pandas as pd
```

Warmups: 2, Iterations : 2

Counter Operations : 10 Million

Skiplist Preset : 1 Millions, Inserts : 1 Million

Btree Preset : 0, Inserts : 1 Million

Machine : Inteli7

```python
def no_sd(v):
    return v if pd.isnull(v) else float(v.split("s")[0])
    
inteli7 = pd.read_csv("intelli7.csv")
inteli7_wo_sd = inteli7.applymap(no_sd)
inteli7_wo_sd.insert(0, 'Domains', range(1, 1 + len(inteli7)))
inteli7_wo_sd['Counter-Seq'] = inteli7_wo_sd['Counter-Seq'][7]
inteli7_wo_sd['Skiplist-Seq'] = inteli7_wo_sd['Skiplist-Seq'][7]
inteli7_wo_sd['Btree-Seq'] = inteli7_wo_sd['Btree-Seq'][7]

counter = inteli7_wo_sd[['Domains', 'Counter-Seq','Counter-Coarse-Grained', 'Counter-Batched']]
counter.plot.line(x="Domains", ylabel="Seconds", grid=True)

skiplist = inteli7_wo_sd[['Domains', 'Skiplist-Seq','Skiplist-Coarse-Grained', 'Skiplist-Batched']]
skiplist.plot.line(x="Domains", ylabel="Seconds", grid=True)

btree = inteli7_wo_sd[['Domains', 'Btree-Seq','Btree-Coarse-Grained', 'Btree-Batched']]
btree.plot.line(x="Domains", ylabel="Seconds", grid=True)
```
